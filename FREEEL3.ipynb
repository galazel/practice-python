```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

csv_path = "/content/drive/MyDrive/APPSSDEDF/winemag-data-130k-v2.csv"

#load dataset
df = pd.read_csv(csv_path)

#view dataset info
print("Head:")
print(df.head())

print("\nTail:")
print(df.tail())

print("\nInfo:")
df.info()

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

#data access
print(type(df))
print(df['description'])
print(df['description'][1])

print()
print(df.loc[0])
print(df.loc[2:3])
print(df.iloc[[0,2]])

print(df.loc[2:5,['taster_name', 'taster_twitter_handle']])
print(df.loc[:,['country','designation','province']])
print(df.tail(3)[['country','designation','province']])

#filter rows
print(df['country'] == 'US')
print(df[(df['country'] == 'US') & (df['points'] >= 99)])
print(df.sort_values(by='price'))
print(df.sort_values(by=['price','country']))

#add columns
df['score_per_dollar'] = 2

#aggregate data
print(df['points'].describe())
print(df['country'].describe())

#group by
print(df.groupby('country')['points'].mean())
print(df.groupby('country')[['points','price']].mean())
print(df.groupby(['province','country'])['points'].mean())
print(df.groupby(['province','country'])['points'].agg(['min','max']))




```
